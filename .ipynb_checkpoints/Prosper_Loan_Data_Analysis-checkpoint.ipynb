## Dataset Overview

This dataset contains data from over 100,000 prosper loans.
Each row contains information about a specific loan that was funded through the Prosper marketplace.

#### Who is prosper?

Prosper is a San Francisco based company in the peer-to-peer lending marketplace in the United States. Since 2005, Prosper has facilitated more than $15 billion in loans to more than 930,000 people.

#### What is peer2peer lending?

For those who are not familiar, Prosper does not actually lend money, they essentially provide a marketplace for lenders and borrowers to meet, think online dating for loans. As a prospective borrower, you would post a loan request to the marketplace and investors can choose to fund your loan.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# Load csv into dataframe
df = pd.read_csv('prosper_final.csv')
df.head()
```

## APR

Before we introduce other variables I believe it's important to look at APR an isolation to understand what might be considered normal. APR is almost normally distributed, however it does have a slight right skew. With a large amount of loans in the 0.35 - 0.375 range. Our average borrower on Prosper is paying 21.96%

```python
#Dictionary to alter font
t_font = {'weight' : 'bold',
        'size'   : 18}
        
l_font ={'weight' : 'bold',
        'size'   : 14}
        

# Using Hist to plot our distribution for APR
sb.set(style="darkgrid")
bins = np.arange(0.05, 0.45, 0.025)
plt.figure(figsize=(16, 12))
plt.hist(data = df, x = 'BorrowerAPR', bins=bins)
plt.title('Distribution of Loan Annual Percentage Rate (APR)', fontdict=t_font)
plt.xlabel('Loan APR', fontdict=l_font, labelpad=10)
plt.ylabel('Loans (count)', fontdict=l_font, labelpad=2);
```

## Borrower APR by Category

If we think of APR as a proxy to sentiment of inherent risk as viewed by our investors, our box plot demonstrates personal loans to be the lest risky based on median 17.6%, Cosmetic Procedure seems to be the one that carries the highest risk with a median of 27.2%

```python
# Groupby Listing Category and return the median
listingCat_median_APR = df.groupby('ListingCategory', as_index=False)['BorrowerAPR'].median()
# Sort by Borrower APR
listingCat_median_APR_val = listingCat_median_APR.sort_values(by='BorrowerAPR')
listingCat_median_APR = list(listingCat_median_APR_val['ListingCategory'])
#Create boxplot
ax = sb.boxplot(data = df, x = 'BorrowerAPR', y = 'ListingCategory', color="royalblue", order=listingCat_median_APR);
plt.gcf().set_size_inches(20, 16);
plt.title('Borrower APR by Category', fontdict=t_font);
plt.xlabel('Borrower APR',fontdict=l_font, labelpad=10)
plt.ylabel('Category',fontdict=l_font, labelpad=10);

```

## Risk Models Over Time

Credit Score and Prosper Score have the strongest negative correlation of all variables -49.9% for the former and -66.8% for the latter. Said another way as your score of either proprietary model increases as the expectancy that your APR will drop increases. Here are some key insights:

### Credit Score Impact on APR Over Time

- Best rates (below 10%) are reserved for the best Credit scores (800+)
- Post 2008 Prosper increased their credit score requirement, the yellow dots have disappeared
- 2011 onwards max APR has dropped from above 40% to slightly above 36%
- 2011 onwards number of borrowers with 800+ credit rating is trending downwards

### Prosper Score Impact on APR Over Time

- Best rates (below 10%) are reserved for the best Prosper Scores (10+)
- Appetite for risk has increased over time
- 2013 - 2014 Prosper Score follows a gradual transition from purple to yellow, investors seemed to have really adopted this risk rating
- 2011 onwards max APR has dropped from above 40% to slightly above 36%


```python
#Dictionary to alter font
t_font = {'weight' : 'bold',
        'size'   : 18}
l_font ={'weight' : 'bold',
        'size'   : 16}


# Scatter plot to show relationship between Creditscore, ProsperScore and APR
plt.subplot(1, 2, 1)
plt.gcf().set_size_inches(20, 10)
plt.subplots_adjust(wspace=0.3)
plt.scatter(data=df, x='Year', y='BorrowerAPR', c='CreditScoreMean', cmap='viridis_r', alpha=0.2);
cbar = plt.colorbar()
cbar.solids.set(alpha=1)
plt.title('Credit Score Impact on Price Over Time',fontdict=t_font);
plt.xlabel('Year',fontdict=l_font)
plt.ylabel('BorrowerAPR',fontdict=l_font);
plt.yticks(np.arange(0, 0.5, step=0.05));
cbar.set_label(label='Credit Score',size=18,weight='bold');


plt.subplot(1, 2, 2)
plt.gcf().set_size_inches(20, 10)
plt.subplots_adjust(wspace=0.2)
plt.scatter(data=df, x='Year', y='BorrowerAPR', c='ProsperScore', cmap='viridis_r', alpha=0.2);
cbar = plt.colorbar()
cbar.solids.set(alpha=1)
plt.title('Prosper Score Impact on APR Over Time',fontdict=t_font);
plt.xlabel('Year',fontdict=l_font)
plt.ylabel('BorrowerAPR',fontdict=l_font);
plt.yticks(np.arange(0, 0.5, step=0.05));
cbar.set_label(label='ProsperScore',size=18,weight='bold')
```
