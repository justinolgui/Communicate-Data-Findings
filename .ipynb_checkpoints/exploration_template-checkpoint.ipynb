```python
#We'll check to see we got all the right information
Condensed_prosper
```

```python
Condensed_prosper.info()
```

```python
#convert to datetime
Condensed_prosper['ListingCreationDate'] = pd.to_datetime(Condensed_prosper.ListingCreationDate)
```

```python
# Create a new series and then convert to Quarters
quarters = pd.Series(Condensed_prosper['ListingCreationDate'])
quarters = pd.PeriodIndex(quarters, freq='Q')
```

```python
# Add a new column to our DF
Condensed_prosper['Quarters'] = quarters
```

```python
# Grab the unique values in our quarters column
quarters = Condensed_prosper['Quarters'].unique()
```

```python
# Sort our list in chronological order
quarters = sorted(quarters)
```

```python
#Draw our distribution
plt.figure(figsize=(12, 9))
sb.set(style="darkgrid")
sb.set(font_scale=1.12)
sb.countplot(x='Quarters', data=Condensed_prosper, color="blue", order=quarters);
plt.xticks(rotation=50);
plt.title('Loan Origination by Quarter');
```

Aside from the Q4 2012 and the Q1 2013 prosper loans have increased quarter over quarter since 2010. What I find to be very interesting is that Q3 of 2008 prosper loans essentially fell off a cliff. The timing is peculiar because this is when the economic crisis of 2008 came to light. Prosper was clearly hit hard by this economic downturn. The next thing i'll be reviewing is credit grade to inspect the distribution of prosper's borrowers.

```python
#Use value counts to see our values in Credit Grade
Condensed_prosper['CreditGrade'].value_counts()
```

```python
# Replace all nan's with the value "None"
Condensed_prosper['CreditGrade'] = Condensed_prosper['CreditGrade'].replace(np.nan, "None")
```

```python
#Use value counts to see our values in Credit Grade
Condensed_prosper['CreditGrade'].value_counts()
```

```python
# Removing all rows that have No credit rating
creditgrade = Condensed_prosper[Condensed_prosper.CreditGrade != "None"]
```

```python
# Checking to see that we have in fact correctly removed those rows
creditgrade.CreditGrade.value_counts()
```

```python
#Drawing our distribution for CreditGrade
grade = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR', 'NC']
plt.figure(figsize=(4, 6))
sb.set(style="darkgrid")
sb.set(font_scale=1.12)
sb.countplot(x='CreditGrade', data=creditgrade, color="blue", order=grade);
plt.title('CreditGrade');
```

This plot is for the most part normally distributed with the highest number of ratings falling in the C category. We have a much lower sample of Credit Grades as this data was only captured pre 2009 Loans. Let's now see how this compares to the prosper score

```python
Condensed_prosper['ProsperScore'].value_counts()
```

```python
#Drawing our distribution for ProsperScore
plt.figure(figsize=(5, 6))
sb.set(style="darkgrid")
sb.set(font_scale=1.12)
sb.countplot(x='ProsperScore', data=Condensed_prosper, color="blue");
plt.title('ProsperScore');
```

This prosperscore is quite interesting. The mode of the distribution is a 4 which is quite low, however we have a high number of borrowers in the 7 and 9 range. We'll now explore the Term variable to see how long are our borrowers taking out loans for. 

```python
# Using Hist to plot our distribution for term
plt.hist(data = Condensed_prosper, x = 'Term');
plt.title('Loan Term')
plt.xlabel('Number of Months')
plt.ylabel('count');
```

Prosper loans are given in 12, 36 and 60 month terms. It's clear that the majority of our borrowers are opting for a 36 month term. Next I would like to find out what kind of interest rates are these borrowers agreeing to.

```python
# Let's use describe so we can find out what might be the best choice for our bins
Condensed_prosper['MonthlyLoanPayment'].describe()
```

```python
# Using Hist to plot our distribution for MonthlyLoanPayment
bins = np.arange(1, 1000, 65)
plt.figure(figsize=(6, 6))
plt.hist(data = Condensed_prosper, x = 'MonthlyLoanPayment', bins=bins);
plt.title('Distribution of Monthly Loans')
plt.xlabel('MonthlyLoanPayment')
plt.ylabel('count');
```

As demonstrated by our plot the majority of the monthly loan payments seem to be just below $300. Let's see what our borrowers seem to make in monthly income.

```python
Condensed_prosper['StatedMonthlyIncome'].describe()
```

```python
# Using Hist to plot our distribution for StatedMonthlyIncome
bins = np.arange(1, 15000, 500)
plt.figure(figsize=(8, 8))
plt.hist(data = Condensed_prosper, x = 'StatedMonthlyIncome',bins=bins)
plt.title('Distribution of Borrowers Income')
plt.xlabel('Stated Monthly Income')
plt.ylabel('count');
```

It appears as thouugh most of our borrowers have a monthly income of roughly 5,000. This distribution is right skewed with some long tails. Kind of interesting that some of our borrowers have a monthly income over 10K and are choosing to turn towards the P2P lending market. This is self reported income though so who knows how accurate this is. Considering most of the monthly loan payments are around $300 this shouldn't put too much stress on our borrowers. However without knowing what other expenses this borrowers have it's hard to say that with certainty. Next I'd like to explore APR to see how much interest our borrowers are paying. 

```python
#Let's use describe to see how we may place our bins.
Condensed_prosper['BorrowerAPR'].describe()
```

```python
# Using Hist to plot our distribution for APR
bins = np.arange(0.05, 0.45, 0.025)
plt.figure(figsize=(7, 7))
plt.hist(data = Condensed_prosper, x = 'BorrowerAPR', bins=bins)
plt.title('Distribution of Borrowers Annual Percentage Rate (APR)')
plt.xlabel('Borrower APR')
plt.ylabel('count');
```

Our distribution here is almost normally distributed, however it does have a slight right skew. It's absolutely mind boggling that some of these borrowers are paying upwards of 35%. These rates are worse than pretty much every credit card there is on the market. These people must be in some serious financial trouble. Next let's explore the debt to income ratio to see just how bad it is. 

```python
#Let's use describe to see how we may place our bins.
Condensed_prosper['BorrowerAPR'].describe()
```

```python
# Using Hist to plot our distribution for APR
bins = np.arange(0.05, 0.55, 0.04)
plt.figure(figsize=(7, 7))
plt.hist(data = Condensed_prosper, x = 'DebtToIncomeRatio', bins=bins)
plt.title('Distribution of Borrowers Debt to Income Ratio')
plt.xlabel('Debt to Income Ratio')
plt.ylabel('count');
```

This distribution is once again right skewed like the last few we have explored. This is quite surprising to be honest. Based on standard finance metrics the majority of these people seem to be in pretty good financial health. Typically if you have a DTI below 0.36 it's considered to be quite good. The mode of our distribution is below 0.2, I'm quite curious to know why these people are turning to the P2P markets. Could this be the result of strict lending criteria, a lack of financial literacy or is the typical loan approval process just too onerous for our borrowers? Next let's explore how much are these borrowers carrying in revolving balances.

```python
#Let's use describe to see how we may place our bins.
Condensed_prosper['RevolvingCreditBalance'].describe()
```

```python
# Using Hist to plot our distribution for APR
bins = np.arange(0, 80000, 5000)
plt.figure(figsize=(7, 7))
plt.hist(data = Condensed_prosper, x = 'RevolvingCreditBalance', bins=bins)
plt.title('Distribution of Revolving Credit Balance')
plt.xlabel('Revolving Credit Bal')
plt.ylabel('count');
```

Once again, yet another heavily right skewed distribution. It appears as though most of our borrowers have less than 5,000 in revolving credit balances. Let's find out where our borrowers are located.
